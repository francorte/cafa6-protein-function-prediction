## Antes de empezar el an√°lisis: revisi√≥n de los datos

Antes de analizar los alimentos y sacar conclusiones nutricionales, es imprescindible
**revisar c√≥mo est√°n los datos**. Este paso evita errores y malas interpretaciones.

### 1. ¬øQu√© significa `int64`?

En la salida de `df.info()` aparece el tipo de dato de cada columna.

- **int64** indica que la columna contiene **n√∫meros enteros**.
- En este dataset, se utiliza para valores como:
  - calor√≠as,
  - gramos de grasa,
  - miligramos de sodio,
  - etc.

Que una columna sea `int64` significa que:
- los datos son num√©ricos,
- se pueden ordenar, comparar y calcular,
- y son adecuados para an√°lisis cuantitativos.

---

### 2. ¬øQu√© es un valor nulo?

Un **valor nulo** es un dato que falta y se representa como:
- `NaN` (Not a Number).

Esto puede ocurrir porque:
- el alimento no contiene ese nutriente,
- el dato no se ha medido,
- o la informaci√≥n no est√° disponible.

Un valor nulo **no es lo mismo que un valor cero**:
- `0` significa que el nutriente no est√° presente.
- `NaN` significa que **no sabemos el valor**.

---

### 3. ¬øPor qu√© es importante revisar los valores nulos?

Si no revisamos los valores nulos antes de analizar:

- Las medias pueden salir incorrectas.
- Las comparaciones entre alimentos pueden ser enga√±osas.
- Podemos sacar conclusiones nutricionales err√≥neas.

Por eso, antes de empezar el an√°lisis debemos:
1. Saber **qu√© columnas tienen valores nulos**.
2. Decidir si esos valores se:
   - eliminan,
   - sustituyen,
   - o simplemente se tienen en cuenta al interpretar.

Este paso es equivalente, en nutrici√≥n, a **comprobar si una etiqueta est√° completa antes de usarla**.

```python
# Comprobar cu√°ntos valores nulos hay en cada columna
df.isnull().sum()
```

## Clave pedag√≥gica: pensar antes de analizar

Este paso es fundamental desde el punto de vista formativo.

- **No todos los datasets est√°n ‚Äúperfectos‚Äù**.

## Relaci√≥n entre calor√≠as y grasas

Una vez analizado el aporte energ√©tico total, es importante estudiar
**qu√© nutrientes contribuyen m√°s a esas calor√≠as**.

Las **grasas** aportan **9 kcal por gramo**, m√°s del doble que:
- los hidratos de carbono (4 kcal/g),
- y las prote√≠nas (4 kcal/g).

Por este motivo, los alimentos con alto contenido graso suelen presentar
un **aporte energ√©tico elevado**, incluso aunque la cantidad consumida no sea grande.

En este apartado analizaremos:
- la grasa total,
- la grasa saturada,
- y su relaci√≥n con las calor√≠as.

```python
# Estad√≠sticos b√°sicos de la grasa total
df_sel["Total Fat\n(g)"].describe()
```

## Interpretaci√≥n del contenido en grasa total

Al analizar la grasa total podemos observar que:

- Existen productos con **muy bajo contenido graso**.
- Otros productos presentan **cantidades elevadas de grasa**, lo que influye
directamente en su aporte energ√©tico.
- La variabilidad en el contenido graso explica, en gran parte, las diferencias
en calor√≠as entre productos.

Esto refuerza la idea de que **no solo importa cu√°ntas calor√≠as tiene un alimento,
sino de d√≥nde proceden esas calor√≠as**.

```python
# Productos con mayor contenido en grasa total
df_sel[["Company", "Item", "Total Fat\n(g)", "Calories"]].sort_values(
    by="Total Fat\n(g)", ascending=False
).head(10)
```

## Grasa saturada y salud

Dentro de las grasas, la **grasa saturada** tiene especial relevancia
desde el punto de vista de la salud cardiovascular.

Un consumo elevado y frecuente de grasas saturadas se asocia con:
- aumento del colesterol LDL,
- mayor riesgo cardiovascular.

Por ello, no solo es importante analizar la cantidad total de grasa,
sino **el tipo de grasa presente en el alimento**.

```python
# Productos con mayor contenido en grasa saturada
df_sel[["Company", "Item", "Saturated Fat\n(g)", "Total Fat\n(g)"]].sort_values(
    by="Saturated Fat\n(g)", ascending=False
).head(10)
```

## Conclusi√≥n del an√°lisis de grasas

El an√°lisis de las grasas muestra que:

- Los productos con mayor aporte cal√≥rico suelen tener un **alto contenido en grasa**.

- Desde el punto de vista nutricional, estos alimentos deber√≠an consumirse
